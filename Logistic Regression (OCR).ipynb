```python
# Define Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Gradient descent function for logistic regression
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    cost_history = []
    
    for iteration in range(iterations):
        z = np.dot(X, theta)
        h = sigmoid(z)
        gradient = np.dot(X.T, (h - y)) / m
        theta -= alpha * gradient
        
        cost = (-1 / m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
        cost_history.append(cost)
        
    return theta, cost_history

# Flatten the 16x16 images into feature vectors
trainX2_flattened = trainX2.reshape(trainX2.shape[0], -1)
trainX4_flattened = trainX4.reshape(trainX4.shape[0], -1)
testX2_flattened = testX2.reshape(testX2.shape[0], -1)
testX4_flattened = testX4.reshape(testX4.shape[0], -1)

# Define labels for binary classification (2 vs 4)
class_2 = 2
class_4 = 4

# Create binary labels for training
trainY_binary = (trainY == class_2).astype(int)
trainY_binary[trainY == class_4] = 0

# Hyperparameters
alpha = 0.5  # Learning rate
iterations = 1000  # Number of iterations

# Concatenate intercept term to the feature matrices
intercept_trainX2 = np.ones((trainX2_flattened.shape[0], 1))
intercept_trainX4 = np.ones((trainX4_flattened.shape[0], 1))
intercept_testX2 = np.ones((testX2_flattened.shape[0], 1))
intercept_testX4 = np.ones((testX4_flattened.shape[0], 1))

X_train2 = np.hstack((intercept_trainX2, trainX2_flattened))
X_train4 = np.hstack((intercept_trainX4, trainX4_flattened))
X_test2 = np.hstack((intercept_testX2, testX2_flattened))
X_test4 = np.hstack((intercept_testX4, testX4_flattened))

# Combine the training data for both classes
X_train = np.vstack((X_train2, X_train4))
y_train = np.concatenate((trainY_binary[:250], trainY_binary[250:]))

# Initialize theta randomly between 0 and 1
np.random.seed(0)  # For reproducibility
theta = np.random.rand(X_train.shape[1])

# Perform gradient descent on the combined training data
theta, cost_history = gradient_descent(X_train, y_train, theta, alpha, iterations)

# Plot cost history to check convergence
plt.plot(range(iterations), cost_history)
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Cost and iteration')
plt.show()

# Predict function
def predict(X, theta):
    probabilities = sigmoid(np.dot(X, theta))
    return [class_2 if p >= 0.5 else class_4 for p in probabilities]

# Combine the test data for classes 2 and 4
X_test = np.vstack((X_test2, X_test4))
y_test = np.concatenate((testY[:50], testY[50:]))

# Make predictions
predictions = predict(X_test, theta)

# Evaluate accuracy
accuracy = np.mean(predictions == y_test)
print(f"accuracy: {accuracy * 100:.2f}%")
```
