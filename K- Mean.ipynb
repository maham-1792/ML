```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
greenary_image = cv2.imread('Greenary.jpg')
greenary_image = cv2.cvtColor(greenary_image, cv2.COLOR_BGR2RGB)
plt.imshow(greenary_image)
greenary_image_2d = greenary_image.reshape((-1, 3)).astype(np.uint8)
plt.title('Orginal Image Greenary')
plt.axis('off')
```

```python
k=2
def assign_to_clusters(data, centroids):
    # Calculate Euclidean distance between each data point and each centroid
    distances = np.linalg.norm(data - centroids[:, np.newaxis], axis=2)

    # Assign each point to the cluster with the nearest centroid
    labels = np.argmin(distances, axis=0)

    return labels
def update_centroids(data, labels, k):
    new_centroids = np.array([data[labels == i].mean(axis=0) for i in range(k)])
    return new_centroids

def initialize_centroids(data, k):
    np.random.seed(42)
    centroid_indices = np.random.choice(data.shape[0], k, replace=False)
    return data[centroid_indices]

def k_means_clustering(data, k, max_iterations=100):
    centroids = initialize_centroids(data, k)

    for iteration in range(max_iterations):
        labels = assign_to_clusters(data, centroids)
        new_centroids = update_centroids(data, labels, k)

        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

        # Visualize intermediate results
        if iteration % 10 == 0:
            segmented_image = centroids[labels].reshape(greenary_image.shape)
          

    segmented_image = centroids[labels].reshape(greenary_image.shape)

    # Display the segmented image
    plt.figure(figsize=(6, 6))
    plt.imshow(segmented_image.astype(np.uint8))
    plt.title(f'Segmented Image (K={k})')
    
    plt.axis('off')


k_means_clustering(greenary_image_2d, k)
```
