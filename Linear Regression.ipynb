```python
train_x=np.array(traindata['X'])
train_y=np.array(traindata['R'])
```

```python
test_x=np.array(testdata['X'])
test_y=np.array(testdata['R'])
```

```python
plt.plot(test_x,test_y)
```

```python
mean_x = train_x.mean()
mean_y = train_y.mean()
n = len(train_x)

numerator = sum((train_x - mean_x) * (train_y - mean_y))
denominator = sum((train_x - mean_x) ** 2)

b1 = numerator / denominator
b0 = mean_y - b1 * mean_x
```

```python
coefficients_quad = np.polyfit(train_x, train_y, 2)
```

```python
coefficients_cubic = np.polyfit(train_x, train_y, 3)
```

```python
coefficients_4th = np.polyfit(train_x, train_y, 4)
```

```python
coefficients_5th = np.polyfit(train_x, train_y, 5)
```

```python
coefficients_6th = np.polyfit(train_x, train_y, 6)
```

```python
plt.scatter(train_x, train_y, label='Training Data', marker='+')
plt.plot(train_x, b0 + b1 * train_x, 'red',label='Linear Regression')
plt.legend()
plt.xlabel('X')
plt.ylabel('R')
plt.title('Linear Regression')
```

```python
plt.scatter(train_x, train_y, label='Training Data', marker='+')
plt.plot(train_x, np.polyval(coefficients_quad, train_x),'red', label='Quadratic Regression')
plt.legend()
plt.xlabel('X')
plt.ylabel('R')
plt.title('Quadratic Regression')
```

```python
plt.scatter(train_x, train_y, label='Training Data', marker='+')
plt.plot(train_x, np.polyval(coefficients_cubic, train_x), 'red',label='Cubic Regression')
plt.legend()
plt.xlabel('X')
plt.ylabel('R')
plt.title('Cubic Regression')
```

```python
plt.scatter(train_x, train_y, label='Training Data', marker='+')
plt.plot(train_x, np.polyval(coefficients_4th, train_x), 'red',label='Fourth Degree Polynomial')
plt.legend()
plt.xlabel('X')
plt.ylabel('R')
plt.title('Fourth Degree Polynomial Regression')
```

```python
plt.scatter(train_x, train_y, label='Training Data', marker='+')
plt.plot(train_x, np.polyval(coefficients_5th, train_x),'red', label='Fifth Degree Polynomial')
plt.legend()
plt.xlabel('X')
plt.ylabel('R')
plt.title('Fifth Degree Polynomial Regression')
```

```python
plt.scatter(train_x, train_y, label='Training Data', marker='+')
plt.plot(train_x, np.polyval(coefficients_6th, train_x),'red', label='Sixth Degree Polynomial')
plt.legend()
plt.xlabel('X')
plt.ylabel('R')
plt.title('Sixth Degree Polynomial Regression')
```

```python
plt.tight_layout()
plt.show()
```

```python
def calculate_mse(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()
linear_predictions = b0 + b1 * train_x
quad_predictions = np.polyval(coefficients_quad, train_x)
cubic_predictions = np.polyval(coefficients_cubic, train_x)
fourth_predictions = np.polyval(coefficients_4th, train_x)
fifth_predictions = np.polyval(coefficients_5th, train_x)
sixth_predictions = np.polyval(coefficients_6th, train_x)

mse_linear = calculate_mse(train_y, linear_predictions)
mse_quad = calculate_mse(train_y, quad_predictions)
mse_cubic = calculate_mse(train_y, cubic_predictions)
mse_4th = calculate_mse(train_y, fourth_predictions)
mse_5th = calculate_mse(train_y, fifth_predictions)
mse_6th = calculate_mse(train_y, sixth_predictions)

print("Mean Square Error (Linear):", mse_linear)
print("Mean Square Error (Quadratic):", mse_quad)
print("Mean Square Error (Cubic): " ,mse_cubic)
print("Mean Square Error (Fourth Degree): ",mse_4th)
print("Mean Square Error (Fifth Degree): ",mse_5th)
print("Mean Square Error (Sixth Degree): ",mse_6th)
```
